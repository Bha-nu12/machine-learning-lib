```python
# mse=sum(yi-y^)**2//n
# rmse=rootof(mse)
# mae=sum|(yi-y^)|//n


# when to  use mse?
# model training
# model evaluation
# it squuares the errors so we cant use this
# when thhe target column doesnt have outliers


# when to use mae?
# when we have outliers


# mse: it is a regression technique which tells how well model captures the patterns
# above 70 %-model performance is good
# below 20% -model performance is bad

# varience tells risk factor of data but std tells actual comparision to data


# for high range columns(salary) -->use std(here varienece will become so high due  to high range of data)
# for low range columns(experience) -->use varience(here std will become very low due to low range of data)


# conclusion
# 1]r**2 score
# 2] mae and rmse compare it 

# why mae and rmse are compared and why not mse?

#     because mae and rmse are in the same units as the target variable, making them more interpretable than mse.

# mae tells actual truth(avg of actual values)
# rmse is risk factor of data
# always rmse>mae
# if rmse is almost equal to mae then we can say that there are no outliers in the data and model is performing well
# if rmse is much greater than mae then we can say that there are outliers in the data and model is not performing well
# means model is not capturing the patterns of outliers data well

# r**2 score tells performance
```
